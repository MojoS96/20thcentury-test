## Virtual Environments in Python + Apply () & Lambda Functions

```python
#library importing
import pandas as pd
import numpy as np
```

```python
#Random data generation

data = pd.DataFrame({'City' : ['London', 'Rome', 'Berlin', 'Tokyo', 'Seoul', 'Los Angeles', 'Vancouver'],
                    'Region' : ['Europe', 'Europe', 'Europe', 'Asia', 'Asia', 'North America','North America'],
                    'Flight_Prices' : [50,80,80,250,400,350,300],
                    'Desire_to_Visit' : [8,8,7,10,10,4,8]})
                    
data
```

```python
# Step 6 - applying a mathematic lambda function
data['Family_return_flights_cost'] = data['Flight_Prices'].apply(lambda x : x*4)

data
```

```python
#step 7 - Applying a flag or 'label' column using lambda
data['likelihood_of_going'] = data.apply(
    lambda row: "Very Likely" if row['Desire_to_Visit'] >= 7 and row['Family_return_flights_cost'] <1000 
    else ( "Likely" if row['Desire_to_Visit'] >= 7 and row ['Family_return_flights_cost'] >=1000 
    else "Unlikely"), axis=1)
                         
data
```

```python
data.sort_values(by='Desire_to_Visit',ascending=False)
```
